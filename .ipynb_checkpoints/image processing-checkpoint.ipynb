```python
df=pd.read_csv("../input/deeper1/train_rotfaces/train.truth.csv")
columns=["fn", "label"]
```

```python
datagen = ImageDataGenerator(rescale=1./255.,   validation_split=0.20)
test_datagen = ImageDataGenerator(rescale=1./255.)



train_generator= datagen.flow_from_dataframe(
    dataframe=df,
    directory="../input/deeper1/train_rotfaces/train",
    x_col="fn",
    y_col=columns[1],
    batch_size=32,
    seed=42,
    classes = ['rotated_left', 'rotated_right', 'upright', 'upside_down'],
    shuffle=False,
    class_mode="categorical",
    target_size=(150,150),
    subset='training')

valid_generator= datagen.flow_from_dataframe(
    dataframe = df,
    directory="../input/deeper1/train_rotfaces/train",
    x_col="fn",
    y_col=columns[1],
    batch_size=32,
    seed=42,
    shuffle=False,
    classes = ['rotated_left', 'rotated_right', 'upright', 'upside_down'],
    class_mode="categorical",
    target_size=(150,150),
    subset='validation', )



```

```python
disc_name = train_generator.class_indices
```

```python
paths = train_generator.filepaths
labels = train_generator.classes

images = []
for path in paths[:10]:
        img_path = path
        img = image.load_img(img_path, target_size=(150,150))
        x = image.img_to_array(img)
        images.append(x/255.)

for I, n  in zip (images , df["label"].head(10).values):
    print(n )
    plt.imshow(I)
    plt.show()
```

```python
df_test = pd.DataFrame(test_names, columns=["fn"])
```

```python
df_test.columns
```

```python
test_datagen = ImageDataGenerator(rescale=1.0/255)
test_generator = test_datagen.flow_from_dataframe(
        dataframe = df_test,
        directory='../input/deeper-test/test_rotfaces/test/',
        class_mode=None,
        x_col="fn",
        target_size=(150,150))
```

```python

def create_model():
    inception = VGG16(include_top=False, input_shape=(150, 150, 3))
    
    model = Sequential()
    model.add(inception)
    model.add(BatchNormalization())
    model.add(Flatten())
    model.add(Dense(256, activation = "relu"))
    model.add(Dense(4, activation='softmax'))   
    model.summary()
    sgd = SGD(lr=1e-3, decay=1e-6, momentum=0.9, nesterov=True)
    #rmsprop = RMSprop(lr=1e-3, decay=1e-6)
    model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['acc'])
    return model
```

```python
model = create_model()
```

```python

n_train = 39117
batch_size = 300
n_valid = 9779
history = model.fit_generator(
       train_generator,
       steps_per_epoch=n_train//batch_size,
       epochs=2,
       validation_data=valid_generator,
       validation_steps=n_valid//batch_size)
```

```python
y_pred =  model.predict_generator(test_generator, steps = (5361 // 32) + 1,  max_queue_size=10)
y_pred = np.argmax(y_pred, axis=1)
```

```python
paths = test_generator.filepaths
labels = ['rotated_left', 'rotated_right', 'upright', 'upside_down']
```

```python
images = []
for path in paths[:10]:
        img_path = path
        img = image.load_img(img_path, target_size=(150,150))
        x = image.img_to_array(img)
        images.append(x/255.)

for I, y  in zip (images,y_pred[:10] ):
    print(labels[y])
    plt.imshow(I)
    plt.show()
```
